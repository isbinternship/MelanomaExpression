```python
# filtering data

# FROM BEFORE
#importing various things
import miner2
import miner2.preprocess, miner2.coexpression, miner2.mechanistic_inference, miner2.miner, miner2.__init__
import seaborn as sb
import matplotlib
import matplotlib.pyplot as plt
import numpy
import pandas
# variables
GDC_download_folder = '/Users/user/projects/melanoma/GDC_Data/gdc_download/'
results_dir = '/Users/user/projects/melanoma/GDC_Data/results/'

# from creating expression table
expressionDataRaw  = miner2.miner.readExpressionFromGZipFiles(GDC_download_folder)
expressionData = miner2.miner.transformFPKM(expressionDataRaw,fpkm_threshold=1,minFractionAboveThreshold=0.05,highlyExpressed=False)

# coefficient of Variation
def findCOV(expressionData):
    
    zero = numpy.percentile(expressionData,0)
    meanCheck = numpy.mean(expressionData[expressionData>zero].mean(axis=1,skipna=True))
    if meanCheck<0.1:
        return expressionData
    means = expressionData.mean(axis=1,skipna=True)
    stds = expressionData.std(axis=1,skipna=True)
    try:
        transform = (stds/means).T
    except:
        passIndex = numpy.where(stds>0)[0]
        transform = ((expressionData.iloc[passIndex,:].T - means[passIndex])/stds[passIndex]).T
    print("completed z-transformation.")
    return transform

# entropy
def findEntropy(expressionData):
    entropyData = expressionData.apply(miner2.miner.entropy,axis=1)
    return entropyData

# expressionData = miner2.miner.transformFPKM(expressionDataRaw,fpkm_threshold=1,minFractionAboveThreshold=0.5,highlyExpressed=False)
coefficientsOfVariation = findCOV(expressionData)
entropyValues = findEntropy(expressionData)

print(coefficientsOfVariation)
print(entropyValues)
```
